### Analysis
* As expected, cities with the warmest temperatures are closest to the equator (0 degrees latitude).  The tempertures decreased as latitude increased.  There were a handful of cities within twenty degrees of the equator that fell below 40 degrees, but the majority were 40 degrees and above at this time.  
* The cities with the highest wind speeds were found in the northern hemisphere.
* Humidity and cloudiness were evenly distributed between cities in both northern and southern hemispheres.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import datetime
from ratelimit import limits, sleep_and_retry

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "Output/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
weather_df = pd.DataFrame()

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="

# Set up DataFrame
column_names = ["Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp","Wind Speed"]
weather_df = pd.DataFrame(index=cities, columns=column_names)

# Set up Datestamp 
current_date = datetime.datetime.today().strftime('%m-%d-%Y')

# Set limits for calling API
minute = 60
@sleep_and_retry
@limits(calls=50, period=minute)

# Function to get weather data and save to a DataFrame
def call_api(url):
    response = requests.get(url).json()
    weather_df.loc[city, "Cloudiness"] = response["clouds"]["all"]
    weather_df.loc[city, "Country"] = response["sys"]["country"]
    weather_df.loc[city, "Date"] = response["dt"]
    weather_df.loc[city, "Humidity"] = response["main"]["humidity"]
    weather_df.loc[city, "Lat"] = response["coord"]["lat"]
    weather_df.loc[city, "Lng"] = response["coord"]["lon"]
    weather_df.loc[city, "Max Temp"] = response["main"]["temp_max"]
    weather_df.loc[city, "Wind Speed"] = response["wind"]["speed"]
    print(f"Processing Record {idx} | {city}")
    return

# Get weather data for cities
for idx, city in enumerate(cities):
    try:
        call_api(query_url + city)
    except:
        print("City not found. Skipping...")
```

### Export the data to a .csv and display the DataFrame count

```python
weather_df.to_csv(output_data_file, header=True)
weather_df.count()
```
